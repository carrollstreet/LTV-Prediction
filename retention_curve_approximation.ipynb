```python
def hyperbolic_func(X, params):
    A, B, C, D, E = params
    return A / (np.log(X + E)**D + B) + C

def mse(params):
    retention_pred = hyperbolic_func(x, params)
    return np.mean((retention - retention_pred)**2)
```

```python
np.random.seed(42)

initial_params = np.random.rand(5)
print('Initial params:',initial_params)
print('Initial MSE:', mse(initial_params))
```

```python
bfgs = minimize(fun=mse, x0=initial_params)

print(bfgs.message)
if bfgs.success == True:
    print('BFGS MSE:',mse(bfgs.x))
```

```python
class AdaGradOpt:
    
    def __init__(self, learning_rate=0.01, epsilon=1e-8, n_iterations=10000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.epsilon = epsilon
        
    def optimize(self, fun, x0, progress_bar=False):
        x0 = np.asarray(x0).copy()
        G = np.zeros_like(x0)
        
        
        def gradient(fun, x0): 
            h = np.cbrt(np.finfo(float).eps)
            grad = np.zeros_like(x0)
            for i in range(len(x0)):
                x_forward, x_backward = x0.copy(), x0.copy()
                x_forward[i] += h
                x_backward[i] -= h
                grad[i] = (fun(x_forward) - fun(x_backward)) / (2 * h)
            return grad
        
        rng = tqdm(range(self.n_iterations)) if progress_bar else range(self.n_iterations)

        for _ in rng:
            grad = gradient(fun, x0)
            G += grad**2
            x0 -= self.learning_rate * grad / (G**.5 + self.epsilon) 
            
        return x0
```

```python
ago = AdaGradOpt(learning_rate=0.95, n_iterations=10_000)
ago_params = ago.optimize(mse,initial_params,progress_bar=True)
```

```python
print('AdaGrad MSE:',mse(ago_params))
```

```python
np.random.seed(42)

bounds = [(-2, 2), (-2, 2), (-2,2), (-2, 2), (0, 2)]
```

```python
def monte_carlo_opt(fun, bounds, minimize=True, num_samples=1_000_000, random_state=None, progress_bar=False):
    np.random.seed(random_state)
    
    best_params = None
    best_loss = None

    lh = defaultdict(tuple)
    for i in bounds:
        lh['low'] += (i[0],)
        lh['high'] += (i[1],)
    
    params = np.random.uniform(lh['low'], lh['high'], size=(num_samples,len(bounds)))
    
    rng = tqdm(params) if progress_bar else params
    
    for i in rng:
        loss = fun(i)
        
        if minimize:
            if best_loss is None or loss < best_loss:
                best_params = i
                best_loss = loss
        else:
            if best_loss is None or loss > best_loss:
                best_params = i
                best_loss = loss
                
    return best_params, best_loss
```

```python
montecarlo_params, loss = monte_carlo_opt(mse, bounds, num_samples=1_000_000, random_state=42, progress_bar=True)
```

```python
print('Monte-Carlo MSE:',loss)
```
